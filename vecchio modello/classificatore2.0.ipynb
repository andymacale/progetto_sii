```python
def jaccard_similarity(set_A, set_B):
    set_A = set(set_A)
    set_B = set(set_B)
    
    intersezione = len(set_A.intersection(set_B))
    unione = len(set_A.union(set_B))
    ris = 0.0
    if unione != 0.0:
        ris = intersezione / unione
    return ris
```

```python
pivot_diagnosi = diagnosi.groupby(['subject_id', 'hadm_id'])['icd_code'].apply(
    lambda codici: ' '.join(codici.dropna().astype(str))
).reset_index()
pivot_diagnosi.rename(columns={'icd_code': 'clinical_profile'}, inplace=True)

pivot_diagnosi['subject_id'] = pivot_diagnosi['subject_id'].astype(str).str.replace('.0', '', regex=False).str.strip()
pivot_diagnosi['hadm_id'] = pivot_diagnosi['hadm_id'].astype(str).str.replace('.0', '', regex=False).str.strip()

dataset_finale['subject_id'] = dataset_finale['subject_id'].astype(str).str.replace('.0', '', regex=False).str.strip()
dataset_finale['hadm_id'] = dataset_finale['hadm_id'].astype(str).str.replace('.0', '', regex=False).str.strip()

dataset_finale = dataset_finale.merge(pivot_diagnosi, on=['subject_id', 'hadm_id'], how='left')

dataset_finale
```

```python
nltk.download('punkt', quiet=True)
stemmer = SnowballStemmer("english")

def calcola_feature_multimodali(riga):
    testo_completo = str(riga.get('clinical_profile', '')) + " " + str(riga.get('testo_referto', ''))
    testo_pulito = re.sub(r'[^a-zA-Z\s]', ' ', testo_completo.lower())
    tokens = testo_pulito.split()
    feature_set = {stemmer.stem(word) for word in tokens if len(word) > 2}
    
    # Tag esatti per età e sesso
    feature_set.add(f"GENDER_{riga.get('gender', '')}")
    feature_set.add(f"AGE_{riga.get('anchor_age', '')}")
    
    return feature_set
```

```python
print("--- 1. SEPARAZIONE DATI: REFERENCE vs TARGET (Peso) ---")
# Identifichiamo i nulli per il peso
weight_nulli = dataset_finale['Weight'].isna()
df_reference_w = dataset_finale[~weight_nulli]
df_target_w = dataset_finale[weight_nulli]

print(f"Abbiamo {len(df_reference_w)} pazienti completi e {len(df_target_w)} da imputare.")

print("\n--- 2. PRE-CALCOLO N-GRAMMI (Salva-RAM) ---")
# Estraiamo i profili clinici dai reference (quelli che "donano" il peso)
profili_ref_w = df_reference_w.apply(calcola_feature_multimodali, axis=1).tolist()
pesi_ref = df_reference_w['Weight'].tolist()

# Estraiamo i profili dai target (quelli che devono ricevere il peso)
target_sets_w = df_target_w.apply(calcola_feature_multimodali, axis=1).tolist()

# --- 3. FUNZIONE WORKER SNELLITA ---
def trova_peso_leggera(target_set):
    # Rifiuto se non c'è abbastanza sostanza clinica (solo age/gender presenti)
    if len(target_set) <= 2:
        return np.nan
        
    # Calcolo della similarità con la tua funzione jaccard_similarity
    scores = np.array([jaccard_similarity(target_set, profilo) for profilo in profili_ref_w])
    
    # Soglia adattiva (media + 1.5 dev.std)
    soglia = scores.mean() + (1.5 * scores.std())
    gemelli = np.where(scores > soglia)[0]

    if len(gemelli) > 0:
        # Restituiamo la mediana dei pesi dei pazienti più simili
        return np.median([pesi_ref[g] for g in gemelli])
    else:
        return np.nan

print("\n--- 4. ELABORAZIONE A BATCH IN PARALLELO (Anti-Crash) ---")
# Parametri coerenti con quelli usati per la pressione
BATCH_SIZE = 500  
N_JOBS = 2        
imputazioni_peso = []

# Dividiamo la lista di set in blocchi
target_chunks_w = [target_sets_w[i:i + BATCH_SIZE] for i in range(0, len(target_sets_w), BATCH_SIZE)]

for i, chunk_sets in enumerate(target_chunks_w):
    print(f"-> Elaborazione Batch {i+1} di {len(target_chunks_w)} in corso...")
    
    # Usiamo il backend multiprocessing per stabilità
    risultati_batch = Parallel(n_jobs=N_JOBS, backend="multiprocessing")(
        delayed(trova_peso_leggera)(t_set) for t_set in chunk_sets
    )
    
    imputazioni_peso.extend(risultati_batch)

# --- 5. ASSEGNAZIONE E PULIZIA ---
dataset_finale.loc[weight_nulli, 'Weight'] = imputazioni_peso

# Drop dei pazienti che non hanno trovato gemelli clinici validi
righe_prima_w = len(dataset_finale)
dataset_finale.dropna(subset=['Weight'], inplace=True)
righe_dopo_w = len(dataset_finale)
imputati_w = len([i for i in imputazioni_peso if not pd.isna(i)])

print(f"\nPeso: Imputati con successo {imputati_w} valori. Eliminate {righe_prima_w - righe_dopo_w} righe scartate.")
```

```python
print("--- 1. SEPARAZIONE DATI: REFERENCE vs TARGET (Pressione) ---")
val_nulli = dataset_finale['Systolic'].isna() | dataset_finale['Diastolic'].isna()

df_reference_bp = dataset_finale[~val_nulli]
df_target_bp = dataset_finale[val_nulli]

print(f"Abbiamo {len(df_reference_bp)} pazienti completi e {len(df_target_bp)} da imputare.")

print("\n--- 2. PRE-CALCOLO N-GRAMMI (Il trucco salva-RAM) ---")
# Estraiamo TUTTO prima del parallelo. Togliamo Pandas di mezzo.
profili_bp = df_reference_bp.apply(calcola_feature_multimodali, axis=1).tolist()
sys_ref = df_reference_bp['Systolic'].tolist()
dia_ref = df_reference_bp['Diastolic'].tolist()

# Pre-calcoliamo anche i bersagli! 
# target_sets_bp è ora una semplice lista di insiemi di parole.
target_sets_bp = df_target_bp.apply(calcola_feature_multimodali, axis=1).tolist()

# --- 3. FUNZIONE WORKER SNELLITA ---
# Questa funzione ora riceve SOLO un set di parole (peso in RAM: vicinissimo allo zero)
def trova_pressione_leggera(target_set):
    if len(target_set) <= 2:
        return (np.nan, np.nan)

    scores = np.array([jaccard_similarity(target_set, profilo) for profilo in profili_bp])
    soglia = scores.mean() + (1.5 * scores.std())
    gemelli = np.where(scores > soglia)[0]

    if len(gemelli) > 0:
        sys_mediana = np.median([sys_ref[g] for g in gemelli])
        dia_mediana = np.median([dia_ref[g] for g in gemelli])
        
        if sys_mediana > dia_mediana:
            return (sys_mediana, dia_mediana)
            
    return (np.nan, np.nan)

print("\n--- 4. ELABORAZIONE A BATCH IN PARALLELO (Anti-Crash) ---")
risultati_totali = []

# Dividiamo in blocchi la lista leggera di set, NON il dataframe Pandas
target_chunks_bp = [target_sets_bp[i:i + BATCH_SIZE] for i in range(0, len(target_sets_bp), BATCH_SIZE)]

for i, chunk_sets in enumerate(target_chunks_bp):
    print(f"-> Elaborazione Batch {i+1} di {len(target_chunks_bp)} in corso...")
    
    # backend="multiprocessing" è più grezzo ma non va in memory leak facilmente
    risultati_batch = Parallel(n_jobs=N_JOBS, backend="multiprocessing")(
        delayed(trova_pressione_leggera)(t_set) for t_set in chunk_sets
    )
    
    risultati_totali.extend(risultati_batch)

# --- 5. ASSEGNAZIONE E PULIZIA ---
imputazioni_sys = [res[0] for res in risultati_totali]
imputazioni_dia = [res[1] for res in risultati_totali]

dataset_finale.loc[val_nulli, 'Systolic'] = imputazioni_sys
dataset_finale.loc[val_nulli, 'Diastolic'] = imputazioni_dia

righe_prima_bp = len(dataset_finale)
dataset_finale.dropna(subset=['Systolic', 'Diastolic'], inplace=True)
righe_dopo_bp = len(dataset_finale)
imputati_bp = len([i for i in imputazioni_sys if not pd.isna(i)])

print(f"\nPressione: Imputati con successo {imputati_bp} valori. Eliminate {righe_prima_bp - righe_dopo_bp} righe scartate.")
```

```python
from joblib import Parallel, delayed
import numpy as np
import pandas as pd

print("--- 1. SEPARAZIONE DATI: REFERENCE vs TARGET (Laboratori) ---")

# Giustissimo: peschiamo chiunque abbia ALMENO un buco in queste 4 colonne
val_nulli = dataset_finale['Hemoglobin'].isna() | dataset_finale['Lactate'].isna() | \
            dataset_finale['Creatinine'].isna() | dataset_finale['WBC'].isna()

df_reference_lab = dataset_finale[~val_nulli]
df_target_lab = dataset_finale[val_nulli]

print(f"Pazienti completi (Reference): {len(df_reference_lab)}")
print(f"Pazienti con almeno un buco (Target): {len(df_target_lab)}")

print("\n--- 2. PRE-CALCOLO N-GRAMMI E FEATURE ---")
# Estraiamo le feature testuali/cliniche per calcolare la Jaccard
profili_lab = df_reference_lab.apply(calcola_feature_multimodali, axis=1).tolist()

# Liste di riferimento
hem_ref = df_reference_lab['Hemoglobin'].tolist()
lac_ref = df_reference_lab['Lactate'].tolist()
cre_ref = df_reference_lab['Creatinine'].tolist()
wbc_ref = df_reference_lab['WBC'].tolist()

target_sets_lab = df_target_lab.apply(calcola_feature_multimodali, axis=1).tolist()

# --- 3. MOTORE DI RICERCA PAZIENTI SIMILI ---
def trova_lab_buona(target_set):
    if len(target_set) <= 2:
        return (np.nan, np.nan, np.nan, np.nan)

    scores = np.array([jaccard_similarity(target_set, profilo) for profilo in profili_lab])
    soglia = scores.mean() + (1.5 * scores.std())
    gemelli = np.where(scores > soglia)[0]

    if len(gemelli) > 0:
        hem_mediana = np.median([hem_ref[g] for g in gemelli])
        lac_mediana = np.median([lac_ref[g] for g in gemelli])
        cre_mediana = np.median([cre_ref[g] for g in gemelli])
        wbc_mediana = np.median([wbc_ref[g] for g in gemelli])

        return (hem_mediana, lac_mediana, cre_mediana, wbc_mediana)
            
    return (np.nan, np.nan, np.nan, np.nan)

print("\n--- 4. ELABORAZIONE A BATCH IN PARALLELO ---")
risultati_totali = []
BATCH_SIZE = 1000
N_JOBS = -1 

target_chunks_lab = [target_sets_lab[i:i + BATCH_SIZE] for i in range(0, len(target_sets_lab), BATCH_SIZE)]

for i, chunk_sets in enumerate(target_chunks_lab):
    print(f"-> Elaborazione Batch {i+1} di {len(target_chunks_lab)}...")
    risultati_batch = Parallel(n_jobs=N_JOBS, backend="multiprocessing")(
        delayed(trova_lab_buona)(t_set) for t_set in chunk_sets
    )
    risultati_totali.extend(risultati_batch)

print("\n--- 5. ASSEGNAZIONE CHIRURGICA (Solo sui NaN) ---")
# Creiamo un DataFrame temporaneo con le imputazioni calcolate, allineato agli indici del target
df_imputazioni = pd.DataFrame({
    'Hemoglobin': [res[0] for res in risultati_totali],
    'Lactate': [res[1] for res in risultati_totali],
    'Creatinine': [res[2] for res in risultati_totali],
    'WBC': [res[3] for res in risultati_totali]
}, index=df_target_lab.index)

# IL TRUCCO: Usiamo fillna() in modo mirato.
# Sostituisce i NaN del dataset_finale con i valori di df_imputazioni, 
# ma NON tocca i valori reali preesistenti!
dataset_finale['Hemoglobin'] = dataset_finale['Hemoglobin'].fillna(df_imputazioni['Hemoglobin'])
dataset_finale['Lactate'] = dataset_finale['Lactate'].fillna(df_imputazioni['Lactate'])
dataset_finale['Creatinine'] = dataset_finale['Creatinine'].fillna(df_imputazioni['Creatinine'])
dataset_finale['WBC'] = dataset_finale['WBC'].fillna(df_imputazioni['WBC'])

# Check e pulizia finale per chi non ha trovato nessun gemello clinico
righe_prima = len(dataset_finale)
dataset_finale.dropna(subset=['Hemoglobin', 'Lactate', 'Creatinine', 'WBC'], inplace=True)
righe_dopo = len(dataset_finale)

print(f"Laboratori: Dati reali preservati. Buchi riempiti chirurgicamente.")
print(f"Eliminate {righe_prima - righe_dopo} righe scartate (nessun gemello utile trovato).")
```

```python
def semplifica_razza(razza):
    razza = str(razza).upper()
    if 'WHITE' in razza or 'PORTUGUESE' in razza:
        return 'WHITE'
    elif 'BLACK' in razza:
        return 'BLACK'
    elif 'HISPANIC' in razza or 'SOUTH AMERICAN' in razza:
        return 'HISPANIC'
    elif 'ASIAN' in razza:
        return 'ASIAN'
    elif 'UNKNOWN' in razza or 'DECLINED' in razza or 'OBTAIN' in razza:
        return 'UNKNOWN'
    else:
        return 'OTHER'

if 'race' in dataset_finale.columns:
    dataset_finale['race'] = dataset_finale['race'].apply(semplifica_razza)
```

```python
if 'language' in dataset_finale.columns:
    dataset_finale['language'] = np.where(dataset_finale['language'] == 'English', 'English', 'Other')
```

```python
colonne_id = ['dicom_id', 'subject_id', 'hadm_id']
id_presenti = [c for c in colonne_id if c in dataset_finale.columns]

if id_presenti:
    dataset_finale = dataset_finale.set_index(id_presenti)
```

```python
def semplifica_dimissione(dimissione):
    dimissione = str(dimissione).upper()
    
    if 'HOME' in dimissione:
        return 'HOME' # Casa o assistenza domiciliare
    elif dimissione in ['SKILLED NURSING FACILITY', 'REHAB', 'CHRONIC/LONG TERM ACUTE CARE', 'ASSISTED LIVING']:
        return 'POST_ACUTE_CARE' # Strutture di riabilitazione o lungodegenza
    elif dimissione in ['DIED', 'HOSPICE']:
        return 'MORTALITY_HOSPICE' # Decesso o fine vita
    else:
        return 'OTHER_UNKNOWN' # Sconosciuti, contro il parere medico, trasferimenti psichiatrici

if 'discharge_location' in dataset_finale.columns:
    dataset_finale['discharge_location'] = dataset_finale['discharge_location'].apply(semplifica_dimissione)
```

```python
if 'marital_status' in dataset_finale.columns:
    dataset_finale['marital_status'] = dataset_finale['marital_status'].fillna('UNKNOWN')
```

```python
vitali = omr.pivot_table(index='subject_id', columns='result_name', values='result_value', aggfunc='first').reset_index()
vitali
```

```python
vitali.info()
```

```python
vitali['Weight (Lbs)'] = pd.to_numeric(vitali['Weight (Lbs)'], errors='coerce')
vitali['Height (Inches)'] = pd.to_numeric(vitali['Height (Inches)'], errors='coerce')
vitali['Weight (kg)'] = vitali['Weight (Lbs)'] * 0.4536
vitali['Height (m)'] = vitali['Height (Inches)'] * 0.0254
vitali['Blood Pressure'] = pd.to_numeric(
    vitali['Blood Pressure'].astype(str).str.split('/').str[0],
    errors='coerce'
)
vitali
```

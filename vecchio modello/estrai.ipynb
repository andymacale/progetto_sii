```python
lista_immagini = []
ricerca = Path(os.path.join("data", "MIMIC_SUPER_RES_24K"))
for file_path in ricerca.rglob('*.jpg'):
    dicom_id = file_path.stem
    lista_immagini.append({'dicom_id': dicom_id, 'path_immagine': str(file_path)})
immagini = pd.DataFrame(lista_immagini)
immagini['path_immagine'] = immagini['path_immagine'].str.replace(str(path+"/"), "data/", regex=False) # pulisci il path
immagini['dicom_id'] = immagini['dicom_id'].astype(str)
print(f"{len(immagini)} immagini")
immagini.head()
```

```python
dataset = immagini.merge(metadati, how='left', on='dicom_id')
dataset
```

```python
mappa_referti = {}
for _, row in dataset.iterrows():
    s_id = str(row['study_id'])
    p_id = str(row['subject_id'])
    nome_originale = f"s{s_id}.txt"
    nome_nuovo = f"p{p_id}_s{s_id}.txt"
    mappa_referti[nome_originale] = nome_nuovo

print(f"Dobbiamo estrarre esattamente {len(mappa_referti)} referti testuali unici.")

# --- 3. SCANSIONE ED ESTRAZIONE (SENZA GUARDARE LE CARTELLE) ---
cartella_mimic = os.path.join("data", "mimic-cxr-reports")
cartella_destinazione = os.path.join("data", "referti_esatti_24k")
os.makedirs(cartella_destinazione, exist_ok=True)

referti_trovati = []

print("Scansione della cartella madre dei referti in corso...")
for radice, _, file in os.walk(cartella_mimic):
    for nome_file in file:
        # Se il file è esattamente uno di quelli che stiamo cercando:
        if nome_file in mappa_referti:
            percorso_sorgente = os.path.join(radice, nome_file)
            nuovo_nome = mappa_referti[nome_file]
            percorso_destinazione = os.path.join(cartella_destinazione, nuovo_nome)
            
            referti_trovati.append((percorso_sorgente, percorso_destinazione))

print(f"Trovati fisicamente {len(referti_trovati)} referti su {len(mappa_referti)}.")

# --- 4. COPIA DEI FILE ---
if len(referti_trovati) > 0:
    with tqdm(total=len(referti_trovati), desc="Copia Referti") as pbar:
        for sorgente, destinazione in referti_trovati:
            shutil.copy2(sorgente, destinazione)
            pbar.update(1)
            
    print(f"\n✅ Estrazione chirurgica completata! I file perfetti sono in: {cartella_destinazione}")
else:
    print("Errore: Nessun referto trovato. Controlla che il percorso 'data/mimic-cxr-reports' sia corretto!")
```
